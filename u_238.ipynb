```python
import tensorflow as tf 
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np

data=pd.read_csv("/home/ibne_walid/Documents/fif/u_238_scattering_captureOutput.csv",on_bad_lines='skip')

x=data["MeV"]
y=data["barns"]
x=np.log(x)
y=np.log(y)

plt.figure(figsize=(10,6))
plt.plot(x,y,color="red")
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a model using the Sequential API
model = tf.keras.Sequential([
tf.keras.layers.Dense(100), # 100 units
tf.keras.layers.Dense(10), # 10 units
tf.keras.layers.Dense(1)])
# Compile the model
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=["mse"])
# Fit the model

model.fit(tf.expand_dims(x, axis=-1), y, epochs=5)


y_predict=model.predict(x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y_predict,color="green")
plt.plot(x,y,color="red")
```
